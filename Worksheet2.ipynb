Forward Fill: Ideal for datasets with a logical progression where past values are likely to inform missing entries (e.g., medical records or sequential data).

Backward Fill: Useful if future data can logically replace missing values.

Combined: Ensures that all gaps are filled when there's no prior value available (using forward fill) and no future value available (using backward fill).

```python
# Problem 3 -
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

df3 = pd.read_csv("/content/drive/MyDrive/Concept&Tech-AI/Titanic-Dataset.csv")

# Step 1: Initialize the encoder
encoder = OneHotEncoder(sparse_output = False)

# Step 2: Fit and transform the 'Embarked' column
embarked_encoded = encoder.fit_transform(df3[['Embarked']])

# Step 3: Create a DataFrame for the encoded values
encoded_df3 = pd.DataFrame(embarked_encoded, columns=encoder.get_feature_names_out(['Embarked']))

# Step 4: Add the new columns to the original DataFrame
df3 = pd.concat([df3, encoded_df3], axis=1)

# Step 5: Drop the original 'Embarked' column
df3 = df3.drop('Embarked', axis=1)

df3.head()
```

```python
# Problem 4 -
import matplotlib.pyplot as plt

df3 = pd.read_csv("/content/drive/MyDrive/Concept&Tech-AI/Titanic-Dataset.csv")

mean_survival_by_sex = df3.groupby('Sex')['Survived'].mean()

print(mean_survival_by_sex)

mean_survival_by_sex.plot.bar(color=['red'])
plt.xlabel('Gender')
plt.ylabel('Mean Survival')
plt.title('Title: Survival Distribution by Gender')
plt.xticks(rotation = 0)
plt.tight_layout()
plt.show()
```

```python
# Problem 5 -

import pandas as pd
import matplotlib.pyplot as plt

#Load the dataset
df3 = pd.read_csv("/content/drive/MyDrive/Concept&Tech-AI/Titanic-Dataset.csv")

# Group by 'Sex' and 'Embarked' and calculate the mean survival rate
# groupby('sex', 'embarked')('survived') -> data is grouped by sex and embarked, forming multi-level index
# unstack() -> forms 'Sex' rows and 'Embarked' columns
mean_survival_by_gender_embarked = df3.groupby(['Sex', 'Embarked'])['Survived'].mean().unstack()

# Print the mean survival rates
print(mean_survival_by_gender_embarked)

# Visualization: Create a bar plot to show the mean survival rates by gender and embarkation port
mean_survival_by_gender_embarked.plot.bar(color=['blue', 'red', 'green'])

# Customize the plot
plt.title('Mean Survival Rates by Gender and Embarkation Port')
plt.xlabel('Gender')
plt.ylabel('Mean Survival Rate')
plt.xticks(rotation=0)  # Keep the x-axis labels horizontal
plt.legend(title='Embarked', labels=['Cherbourg (C)', 'Queenstown (Q)', 'Southampton (S)'])
plt.show()
```

```python
# Problem 6 -
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df3 = pd.read_csv("/content/drive/MyDrive/Concept&Tech-AI/Titanic-Dataset.csv")

df3['Age'] = df3['Age'].fillna(df3['Age'].median())

# Create 5 quantiles based on 'Age' column
df3['Age Group'] = pd.qcut(df3['Age'], q=5, duplicates = 'drop')

# Check the DataFrame
print(df3[['Age', 'Age Group']].head())

mean_survival_by_age_passenger = df3.groupby(['Age', 'Pclass'])['Survived'].mean().unstack()

# Print the mean survival rates
print(mean_survival_by_age_passenger)

# Visualization: Create a bar plot to show the mean survival rates by gender and embarkation port
# seaborn's barplot -> easier to visualize the relationships between categorical variables and their aggregated values, like the mean, sum, or count.
plt.figure(figsize=(10, 6))
sns.barplot(x='Age Group', y='Survived', hue='Pclass', data=df3, ci=None, palette='muted')

# Customize the plot
plt.title('Mean Survival Rates by Age and Passenger Class')
plt.xlabel('Age')
plt.ylabel('Mean Survival Rate')
plt.xticks(rotation=0)  # Keep the x-axis labels horizontal
plt.show()
```
