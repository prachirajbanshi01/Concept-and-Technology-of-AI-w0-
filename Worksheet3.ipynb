```python
# Missing values in the Age column by filling them with the median age of the dataset

median_age = df["Age"].median()

# Step 2: Fill missing values in the "Age" column with the median
df["Age"] = df["Age"].fillna(median_age)

# Step 3: Perform division (e.g., divide all age values by 2)
df["Age"] = df["Age"] / 2

# Display the updated DataFrame
print("Updated DataFrame by filling Age column with the median age of dataset:")
print(df)
```

```python
# 1 : Add a column 'Fare per year' (Fare/Age)

# fare_per_year = df["Fare"]/(df["Age"])
# print(fare_per_year)

df['Fare per year'] = df['Fare'] / df['Age']
print("Updated DataFrame with new column 'Fare per year': ")
df
```

```python
# 2 : Subset rows with fare-per-year higher than 5

high_fare_age = df.loc[df['Fare per year'] > 5]
print("\nSubset with fare-per-year higher than 5: \n", high_fare_age)
```

```python
# 3 : Sort high_fare_age by descending fare_per_year

high_fare_age_srt = df.sort_values(by = 'Fare per year', ascending=False)
print("\nSorting high-fare-age by descending fare-per-year: \n", high_fare_age_srt)
```

Problem 3.3 : Exploratory Data Analysis with Group-by Method Practice Exercise

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Concept&Tech-AI/Titanic-Dataset.csv")

df
```

**What percent of the total fare revenue came from each passenger class?

```python
# 1 : Calculate Total Fare paid across all passengers

total_fare = df['Fare'].sum()
print("Total Fare: ", total_fare)
```

```python
# 2 : Subset for passengers in first class (Pclass = 1) and calcuate their total fare

first_class_fare = df.loc[df['Pclass'] == 1, 'Fare'].sum()
print("Total Fare for passengers in first class: ", first_class_fare)
```

```python
# 3 : Subset for passengers in second and third class (Pclass = 2 & Pclass = 3)

second_class_fare = df.loc[df['Pclass'] == 2, 'Fare'].sum()
print("Total Fare for passengers in second class: ", second_class_fare)

third_class_fare = df.loc[df['Pclass'] == 3, 'Fare'].sum()
print("Total Fare for passengers in third class: ", third_class_fare)

print("-------------------------------------------------")

second_third_class_fare = df.loc[df['Pclass'].isin([2, 3]), 'Fare'].sum()
print("Total Fare for passengers in second and third class: ", second_third_class_fare)
```

```python
# 4 : Combine fare totals from first, second and third classes into a list

fare_totals = [total_fare, first_class_fare, second_class_fare, third_class_fare]
print("List of fare totals: ", fare_totals)
```

```python
# 5 : Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class

first_proportion = first_class_fare / total_fare
second_proportion = second_class_fare / total_fare
third_proportion = third_class_fare / total_fare

print("--Proportion of fare revenue by class--")
print("First class: ",first_proportion)
print("Second class: ",second_proportion)
print("Third class: ",third_proportion)

# proportion_by_class = [total / total_fare for total in fare_totals]
# print("Proportion of fare revenue by class: ", proportion_by_class)
```

```python
# What percent of the total fare revenue came from each passenger class?

print("Percentage proportion of total fare from each passenger class")

first_class_proportion_percent  = first_proportion * 100
print("First class proportion: ", first_class_proportion_percent, "%")

second_class_proportion_percent  = second_proportion * 100
print("Second class proportion: ", second_class_proportion_percent, "%")

third_class_proportion_percent  = third_proportion * 100
print("Third class proportion: ", third_class_proportion_percent, "%")
```

**What percent of the total number of passengers on the Titanic belonged to each age group
(e.g., child, adult, senior)?

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Concept&Tech-AI/Titanic-Dataset.csv")

df
```

```python
# 1 : Create new column 'Age group' and categorize

# function to categorize the age group
def categorize_age(age):
    if age < 18:
        return "Child"
    elif age < 65:
        return "Adult"
    else:
        return "Senior"

# Create a new column 'age group'
df["Age Group"] = df["Age"].apply(categorize_age)

# Display the updated DataFrame with the new column
print(df[["Age", "Age Group"]])
```

```python
# Updated dataframe with new column 'Age Group'
df
```

```python
# 2 : Calculate total number of passengers

total_passengers = len(df)
print("Total number of passengers: ", total_passengers)
```

```python
# 3 : Count the number of passengers in each age group
age_group_counts = {
    "Child": (df["Age Group"] == "Child").sum(),
    "Adult": (df["Age Group"] == "Adult").sum(),
    "Senior": (df["Age Group"] == "Senior").sum(),
}

print("Number of passengers in each age group: \n", age_group_counts)
```

```python
# 4 : Divide the count of each age group by the total number of passengers

child_count_proportion = age_group_counts['Child'] / total_passengers
print("Child Proportion: ",child_count_proportion)

adult_count_proportion = age_group_counts['Adult'] / total_passengers
print("Adult Proportion: ", adult_count_proportion)

senior_count_proportion = age_group_counts['Senior'] / total_passengers
print("Senior Proportion: ", senior_count_proportion)
```

```python
# 5 : Display proportion as a percentage

child_count_proportion_percent = child_count_proportion * 100
print("Child Proportion: ",child_count_proportion_percent,"%")

adult_count_proportion_percent = adult_count_proportion * 100
print("Adult Proportion: ", adult_count_proportion_percent,"%")

senior_count_proportion_percent = senior_count_proportion * 100
print("Senior Proportion: ", senior_count_proportion_percent,"%")
```
