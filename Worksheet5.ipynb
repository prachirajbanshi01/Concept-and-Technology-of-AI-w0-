```python
# Build a cost function

def cost_function(X, Y, w):
  """Parameters:
  This function finds the Mean Square Error.
  Input parameters:
    X : Feature Matrix ()
    Y : Target Matrix
    w : Weight Matrix
  Output:
    cost : accumulated Mean Square Error
  """
  # number of samples
  m = X.shape[0]
  # predicted values (w^T * X)
  Y_pred = np.dot(X, w)
  # Mean Square Error : (Y-pred(i) - Y(i))^2 ; where Y-pred(w) = w^T * X
  cost = np.sum((Y_pred - Y)**2) / (2 * m)
  return cost

cost_value = cost_function(X, Y, w)
print("Cost Value:", cost_value)
```

```python
from typing import final
# Gradient Descent

def gradient_descent(X, Y, w, alpha, iterations):
  """Perform gradient descent to optimize the parameters of a linear regression model.
  Parameters:
    X : Feature Matrix (m,n)
    Y : Target Matrix (m,1)
    w : Initial guess for parameters(n,1)
    alpha (float) : learning rate
    iterations (int) : number of iterations for gradient descent.
  Returns:
  tuple: A tuple containing the final optimized parameters(W_update) and the history of cost values.
  w_update : Updated parameters (n,1)
  cost_hitory : History of cost values
  """
  cost_history = [0] * iterations
  # Number of samples
  m = len(Y)

  # w_update = initial weight at first
  w_update = w
  for iteration in range(iterations):
    # Step 1: Hypothesis values
    Y_pred = np.dot(X, w_update)

    # Step 2: Difference between Hypothetis and Actual Y
    loss = Y_pred - Y

    # Step 3: Gradient Calculation
    dw = (1 / m) * np.dot(X.T, loss)

    # Step 4: Updating value of w using gradient
    w_update = w_update - alpha * dw

    # Step 5: New cost value
    cost = cost_function(X, Y, w_update)

    cost_history[iteration] = cost

  return w_update, cost_history

# Test case
np.random.seed(0)
X = np.random.rand(100, 3)
Y = np.random.rand(100)
w = np.random.rand(3)

alpha = 0.01 # learning rate
iterations = 1000 # number of iterations

final_parameters, cost_history = gradient_descent(X, Y, w, alpha, iterations)

print("Final Parameters:", final_parameters)
print("Cost History:", cost_history)
```

```python
# Evaluate the model

def rmse(Y , Y_pred):
  """
  This function calculates the Root Mean Squares.
  Input Arguments:
  Y : Array of actual (target) dependent variables
  Y_pred : Array of predicted dependent variables
  Output Arguments:
  rmse : Root Mean Squares
  """
  mse = np.mean((Y - Y_pred) ** 2)
  rmse = np.sqrt(mse)
  return rmse

# Test case
np.random.seed(0)
Y = np.random.rand(100)
Y_pred = np.random.rand(100)  # Simulated predictions

result = rmse(Y, Y_pred)
print("Root Mean Squared Error (RMSE): ", result)
```

```python
# Model evaluation - r2
# here, r2 = 1 - (residual sum of squares / total sum of squares)
# where, residual sum squares = (Y - Y_pred)^2
# and, total sum squares = (Y - Mean_Y)^2

def r2(Y, Y_pred):
  """
  This function calculates the R squared value.
  Input Arguments:
  Y : Array of actual (target) dependent variables
  Y_pred : Array of predicted dependent variables
  Output Arguments:
  r2 : R squared value
  """

  mean_y = np.mean(Y)
  # Total sum of squares
  ss_total = np.sum((Y - mean_y) ** 2)
  # Residual sum of squares
  ss_residual = np.sum((Y - Y_pred) ** 2)
  # R squared value
  r2 = 1 - (ss_residual / ss_total)
  return r2

# Test case
np.random.seed(0)
Y = np.random.rand(100)       # Actual values
Y_pred = np.random.rand(100)  # Predicted values

result = r2(Y, Y_pred)
print("R squared (R2): ", result)
```

```python
# Main function

def main():
  data = pd.read_csv("/content/drive/MyDrive/Worksheet5/student.csv")

  X = data[['Math', 'Reading']].values
  Y = data['Writing'].values

  X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

  w = np.zeros(X.shape[1])
  alpha = 0.00001
  iterations = 1000

  w_optimal, cost_history = gradient_descent(X_train, Y_train, w, alpha, iterations)

  Y_pred = np.dot(X_test, w_optimal)
  Y_train_pred = np.dot(X_train, w_optimal)

  train_rmse = rmse(Y_train, Y_train_pred)
  train_r2 = r2(Y_train, Y_train_pred)

  model_rmse = rmse(Y_test, Y_pred)
  model_r2 = r2(Y_test, Y_pred)


  print("Final Weights: ", w_optimal)
  print("Cost History (First 10 iterations): ", cost_history[:10])
  print("Model RMSE:", model_rmse)
  print("Model R2:", model_r2)

  print("\nTraining RMSE:", train_rmse)
  print("Training R2:", train_r2)


  if model_rmse > train_rmse * 1.2:
    # multiplying it by 1.2 signifies the gap between training data and test data
    print("Model is overfitting")
  elif train_rmse > model_rmse * 1.2:
    print("Model may be underfitting")
  else:
    print("Model performance is acceptable")

if __name__ == "__main__":
  main()


```
